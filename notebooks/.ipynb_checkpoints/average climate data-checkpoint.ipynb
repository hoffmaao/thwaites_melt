```python
import rioxarray
import geopandas as gpd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
geodf = gpd.read_file('/Users/andrew/Downloads/IceShelf_Antarctica_v02.shp')
PIG=geodf[geodf['NAME']=='Pine_Island'].to_crs(epsg=4326)
climate_path='/Volumes/ERA5/data/Pine_Island-iceshelf/'
file_list = os.listdir(climate_path)
```

```python
for file in file_list:
    print(file)
    xds = rioxarray.open_rasterio(climate_path+file_list[0])
    xds.rio.write_crs("epsg:4326", inplace=True)
    clipped = xds.rio.clip(PIG.geometry.values, PIG.crs)*xds.scale_factor+xds.add_offset
    weights = np.cos(np.deg2rad(clipped.y))
    weights.name = "weights"
    clipped_weighted=clipped.weighted(weights)
    clipped_mean = clipped_weighted.mean(("x", "y"))
    df = clipped_mean.to_dataframe()
    df = df.to_csv(climate_path+file[:-2]+'.csv')
```
